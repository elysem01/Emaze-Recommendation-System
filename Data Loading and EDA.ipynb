# Cleaning and Combining Additional Datasets

```python
deletegroups = ['CampaignID', 'CampaignType']
templates = templates_raw.drop(deletegroups, axis=1)
templates.rename(columns={'customThemeID': 'ThemeID'}, inplace=True)
templates.head()
```

```python
themes_raw.columns
```

```python
deletegroups = ['dateEntered', 'thumbnailID', 'domainName', 'supportedBrowsers',
               'themeCssUrl', 'sceneCssUrl', 'sceneJsUrl', 'technology', 'thumbnail',
               'position', 'groupID', 'logoImage', 'tagLine', 'customSettings',
               'DateSaved', 'DatePublished', 'isCustom', 'restriction', 'EzContentTypeId',
               'SocialFacebook', 'SocialInstagram', 'SocialYoutube', 'SocialTwitter']
themes1 = themes_raw.drop(deletegroups, axis=1)
themes1.rename(columns={'themeID': 'ThemeID'}, inplace=True)
themes1.head()
```
