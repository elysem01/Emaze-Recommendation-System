# Importing Essential Data

```python
import numpy as np
import pandas as pd
import seaborn as sns

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
groups_raw = pd.read_csv("data/EmazeMeGroups.csv")
groups_raw.head()
```

```python
followers_raw = pd.read_csv("data/EmazeMeGroupFollowers.csv")
followers_raw.head()
```

```python
members_raw = pd.read_csv("data/EmazeMeGroupMembers.csv")
members_raw.head()
```

```python
pending_raw = pd.read_csv("data/EmazeMeGroupPendingMembers.csv")
pending_raw.head()
```

```python
time_raw = pd.read_csv('data/TimeOnPages.csv')
time_raw.head()
```

# Cleaning and Combining Datasets

```python
members_raw.columns
```

```python
members_raw.dtypes
```

```python
members = members_raw.drop(members_raw.index[:12])
members.head()
```

```python
members.MemberSince = pd.to_datetime(members['MemberSince'])
members.dtypes
```

```python
members.rename(columns={'EmazeMeGroupId': 'GroupID'}, inplace=True)
```

```python
pending = pending_raw
pending.columns
```

```python
pending.rename(columns={'EmazeMeGroupId': 'GroupID'}, inplace=True)
```

```python
groups_raw.columns
```

```python
groups_raw.DateCreated = pd.to_datetime(groups_raw['DateCreated'])
groups_raw.DateUpdated = pd.to_datetime(groups_raw['DateUpdated'])
print(groups_raw.dtypes)
```

```python
deletegroups = ['priority', 'varified', 'AssociatedGroupsJson', 'ModeratorsJson',
                'EmazeSite', 'GroupCoverUrl', 'GroupPhotoUrl', 'SiteThumnail']

groups = groups_raw.drop(deletegroups, axis=1)
groups.rename(columns={'Privecy': 'Privacy'}, inplace=True)
groups.head()
```

```python
time = time_raw[time_raw.PageType == 'group']
time.head()
```

```python
deletegroups = ['Id', 'PageType', 'FirstPing', 'IPAdress']

time = time.drop(deletegroups, axis=1)
time.rename(columns={'Id': 'GroupID', 'PageId': 'GroupName'}, inplace=True)
time.head()
```

```python
data = pd.concat([groups, members, pending, time],ignore_index=True)
data.head()
```

# Importing Additional Data

```python
templates_raw = pd.read_csv('data/EmazeMeGroupFromTemplate.csv')
templates_raw.head()
```

```python
themes_raw = pd.read_csv('data/Themes.csv')
themes_raw.head()
```

```python
categories_raw = pd.read_csv('data/ThemeCategories.csv')
categories_raw.head()
```

```python
names_raw = pd.read_csv('data/ThemeCategoryNames.csv')
names_raw.head()
```

# Cleaning and Combining Additional Datasets

```python
deletegroups = ['CampaignID', 'CampaignType']
templates = templates_raw.drop(deletegroups, axis=1)
templates.rename(columns={'customThemeID': 'ThemeID'}, inplace=True)
templates.head()
```

```python
themes_raw.columns
```

```python
deletegroups = ['dateEntered', 'thumbnailID', 'domainName', 'supportedBrowsers',
               'themeCssUrl', 'sceneCssUrl', 'sceneJsUrl', 'technology', 'thumbnail',
               'position', 'groupID', 'logoImage', 'tagLine', 'customSettings',
               'DateSaved', 'DatePublished', 'isCustom', 'restriction', 'EzContentTypeId',
               'SocialFacebook', 'SocialInstagram', 'SocialYoutube', 'SocialTwitter']
themes1 = themes_raw.drop(deletegroups, axis=1)
themes1.rename(columns={'themeID': 'ThemeID'}, inplace=True)
themes1.head()
```

```python
themes2 = themes1[themes1.isActive == 1]
```

```python
themes = themes2.drop(['isActive'], axis=1)
themes.head()
```

```python
categories = categories_raw
categories.rename(columns={'themeID': 'ThemeID'}, inplace=True)
categories.head()
```

```python
names_raw.head()
```

```python
names1 = names_raw.drop(['position', 'EzContentTypeId'], axis=1)
names1.rename(columns={'name': 'Category'}, inplace=True)
names1.head()
```

```python
names2 = names1[names1.isActive == 1]
```

```python
names = names2.drop(['isActive'], axis=1)
names.head()
```

```python
pd.set_option('display.max_rows', None)
data2 = pd.concat([templates, themes, categories, names],ignore_index=True)
data2
```

# Exploratory Data Analysis for Themes

```python
groupuses = data2.groupby('ThemeID')['GroupID'].count()
groupuses.sort_values(ascending=False).head()

#The greatest number of groups that a theme has been used in is 2
```
